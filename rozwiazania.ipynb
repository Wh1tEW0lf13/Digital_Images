# Zadanie 3

```python
import math
input = 'math_result/input.txt'
output = 'math_result/output.txt'
with open(input, 'r', encoding='utf-8') as plik:
    argumentList = plik.read().splitlines()
with open(output, 'a', encoding='utf-8') as plik:
    for argument in argumentList:
        liczba = float(argument)
        print(f"{liczba} pierwiastek z niej == {math.sqrt(liczba)}")
        plik.write(f"{liczba} pierwiastek z niej == {math.sqrt(liczba)}\n")
```
